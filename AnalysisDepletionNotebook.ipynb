```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-


import openmc
import openmc.deplete
import numpy as np
import matplotlib.pyplot as plt
```

```python
###############################################################################
#                    Read depletion calculation results
###############################################################################
print("Before results list")
# Open results file
# Open results file
results = openmc.deplete.ResultsList("depletion_results.h5")

# Obtain K_eff as a function of time
time, keff = results.get_eigenvalue()

# Obtain U235 concentration as a function of time
time, n_U235 = results.get_atoms('1', 'U235')
time, n_U238 = results.get_atoms('1', 'U238')
# Obtain Xe135 absorption as a function of time
time, Xe_gam = results.get_reaction_rate('1', 'Xe135', '(n,gamma)')
time, n_I135 = results.get_atoms('1', 'I135')
time, n_Cs135 = results.get_atoms('1', 'Cs135')
time, n_Gd157 = results.get_atoms('1', 'Gd157')
time, n_Gd156 = results.get_atoms('1', 'Gd156')

time, n_U234 = results.get_atoms('1', 'U234')

time, n_Th232 = results.get_atoms('1', 'Th232')
time, n_Th233 = results.get_atoms('1', 'Th233')
time, n_Pa233 = results.get_atoms('1', 'Pa233')
time, n_U233 = results.get_atoms('1', 'U233')

#time, fission = results.
```

```python
###############################################################################
#                            Generate plots
###############################################################################

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_Th232, label="Th232")
plt.xlabel("Time (days)")
plt.ylabel("n Th232 (-)")
plt.show()

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_Th233, label="Th233")
plt.xlabel("Time (days)")
plt.ylabel("n Th233 (-)")
plt.show()

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_Pa233, label="Pa233")
plt.xlabel("Time (days)")
plt.ylabel("n Pa233 (-)")
plt.show()

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_U233, label="U233")
plt.xlabel("Time (days)")
plt.ylabel("n Pa233 (-)")
plt.show()

plt.figure()
plt.figure(figsize=(10,10))
plt.plot(time/(24*60*60), keff, label="K-effective")
plt.xlabel("Time (days)")
plt.ylabel("Keff")
plt.show()

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_U235, label="U 235")
plt.xlabel("Time (days)")
plt.ylabel("n U235 (-)")
plt.show()

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), Xe_gam, label="Xe135 absorption")
plt.xlabel("Time (days)")
plt.ylabel("Xe135 absorption")
plt.show()
plt.close('all')

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_I135, label="I135")
plt.xlabel("Time (days)")
plt.ylabel("n I135")
plt.show()

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_Cs135, label="Cs135")
plt.xlabel("Time (days)")
plt.ylabel("n Cs135")
plt.show()

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_Gd157, label="Gd157")
plt.xlabel("Time (days)")
plt.ylabel("n Gd157")
plt.show()
plt.close('all')

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_Gd156, label="Gd156")
plt.xlabel("Time (days)")
plt.ylabel("n Gd156")
plt.show()
plt.close('all')

plt.figure()
plt.figure(figsize=(10,10))
plt.semilogy(time/(24*60*60), n_U234, label="U234")
plt.xlabel("Time (days)")
plt.ylabel("n U234")
plt.show()
plt.close('all')
```
